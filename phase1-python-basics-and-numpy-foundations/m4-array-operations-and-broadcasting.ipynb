In this module, we'll be looking at **Arrays Operations and Broadcasting**<br>
The core idea is simple: NumPy can perform operations on entire arrays without a single loop.

**1 - Element-wise operations**<br>
NumPy applies operations to every element simultaneously.

````python
import numpy as np
````

````python
prices = np.array([100, 105, 102, 110, 108])
````

````python
#These are normal python operations that you can do. The key thing is that they are elementwise. So the operation is applied to each element of the array!
#Thus, no looops!

print(prices) #original array
print("Add 10 to each element:\n", prices + 10) #add 10 to each element
print("Subtract 5 from each element:\n", prices - 5) #subtract 5 from each element
print("Multiply each element by 2:\n", prices * 2) #multiply each element by 2
print("Divide each element by 2:\n", prices / 2) #divide each element by 2
print("Square each element:\n", prices ** 2) #square each element
print("Modulo each element by 3:\n", prices % 3) #modulo each element by 3
````

You can see that under the hood, NumPy is looping BUT it's doing it in C (not python!) Hence why it is faster by orders of magnitude.

**2 - Operations between two arrays**<br>
You can also do these operations BETWEEN two arrays

````python
asset_a = np.array([100, 105, 102, 110, 108])
asset_b = np.array([50,  52,  49,  55,  53 ])

#You can also do operations between arrays. The key thing is that the arrays need to be of the same shape (or compatible shapes, which we will see later). The operation is applied elementwise between the two arrays!
spread = asset_a - asset_b
print("Spread between asset_a and asset_b:\n", spread) #spread between assets, the difference at each point in time

combined = asset_a + asset_b
print("Combined value of asset_a and asset_b:\n", combined) #combined value of assets, the sum at each point in time

ratio = asset_a / asset_b 
print("Ratio of asset_a and asset_b:\n", ratio) #ratio of the two assets, the ratio at each point in time

#You can also do this between multiple arrays, e.g 3
asset_c = np.array([5, 24, 36, 36, 63])

combined2 = asset_a + asset_b + asset_c
print("Combined value of asset_a, asset_b, asset_c:\n", combined2)

````

````python
#You can see here that if they are different shapes, it doesn't work
asset_a2 = np.array([100, 105, 102, 110, 108])
asset_b2 = np.array([[50,  52,  49,  55,  53 ],[5, 24, 36, 36, 63]])
asset_b3 = np.array([[50,  52,  49,  55,  53, 65, 85, 29],[5, 24, 36, 36, 63, 5, 85, 29 ]])

diff_shapes = asset_a2 + asset_b2
print(diff_shapes) # it accepts same shape but but different size

diff_shapes3 = asset_a2 + asset_b3
print(diff_shapes3) # this one should not work!
````

**3 - Calculating Returns**<br>
*Here is a practical example of this ability in practise*

````python
prices3 = np.array([100, 105, 102, 110, 108])

#Absolute daily change
daily_change = np.diff(prices3) #getting the difference between prices
print("Daily change:", daily_change)

# Daily return - this is a standard formula !
# (today - yesterday) / yesterday * 100
daily_returns = daily_change / prices3[:-1] * 100 # we use prices3[:-1] to get all elements except the last one, which corresponds to yesterday's price
print("Daily returns (%):", daily_returns)
````

**4 - Broadcasting**<br>
*THIS is the ability to work on **different** shapes*

````python
#simple broadcasting - scalar and array
a = np.array([1,2,3,4,5])
print(a +10) # add 10 to each element of the array
# 10 is 'broadcasted' to each element of the array, so it is like doing a + 10 for each element of a

#2D broadcasting - this is way more powerful and is the basis of many operations in numpy. The key thing is that the shapes of the arrays need to be compatible. The operation is applied elementwise between the two arrays, but numpy will automatically 'broadcast' the smaller array to match the shape of the larger array.
portfolio = np.array([[100, 105, 102],   # Stock A
                      [50,  52,  49 ],   # Stock B
                      [200, 198, 205]])  # Stock C

#Multiply each stock by a different weight (in one line...)
weights = np.array([0.5, 0.3, 0.2])
weighted_potfolio = portfolio * weights[:, np.newaxis] # we use np.newaxis to add a new axis to the weights array, so it becomes a column vector and can be broadcasted to match the shape of the portfolio array
print(weighted_potfolio) # this gives us the weighted portfolio, where each stock is multiplied by its corresponding weight
````

np.newaxis reshapes weights from shape (3,) to (3,1) so it broadcasts correctly across columns.

remember: (row, column)

so it effectly turns into:

[0.5, 0.5, 0.5],   
[0.3, 0.3, 0.3],   
[0.2, 0.2, 0.2]

**5 - Aggregations Across Axis<br>**
*With 2D arrays you can choose to aggregate across rows or columns*

````python
portfolio = np.array([[100, 105, 102, 110],   # Stock A — 4 days
                      [50,  52,  49,  55 ],   # Stock B
                      [200, 198, 205, 210]])  # Stock C

print(np.mean(portfolio))           # mean of ALL values
print(np.mean(portfolio, axis=0))   # mean of each DAY (across stocks)
    # i.e, we are taking the mean for each column (day)
print(np.mean(portfolio, axis=1))   # mean of each STOCK (across days)
    # i.e, we are taking the mean for each row (stock)
````

Axis 0 = down the rows, Axis 1 = across the columns. 
Remember "axis 0 collapses rows, axis 1 collapses columns"

portfolio = [[100, 105, 102, 110],   ← row 0 (Stock A)
             [50,  52,  49,  55 ],   ← row 1 (Stock B)
             [200, 198, 205, 210]]   ← row 2 (Stock C)
               ↑    ↑    ↑    ↑
              col0 col1 col2 col3
             Day1 Day2 Day3 Day4
            
axis=0 means "travel downward ↓ through rows"
axis=1 means "travel sideways → through columns"

np.mean(portfolio, axis=0)  # collapse the ROWS — one answer per COLUMN
# → "what's the average price on each day?"
# → shape goes from (3,4) to (4,)

np.mean(portfolio, axis=1)  # collapse the COLUMNS — one answer per ROW
# → "what's the average price of each stock?"
# → shape goes from (3,4) to (3,)
```

## The Practical Way 🏦

Just ask yourself:
> *"What do I want ONE answer per ___?"*
```
One answer per STOCK?  → axis=1  (collapse across days)
One answer per DAY?    → axis=0  (collapse across stocks)



````python
pythonportfolio = np.array([[100, 105, 102, 110],
                            [50,  52,  49,  55 ],
                            [200, 198, 205, 210]])

print("No axis — one number for everything:")
print(np.mean(portfolio))           # 118.08...

print("\naxis=0 — one number PER COLUMN (per day):")
print(np.mean(portfolio, axis=0))   # [116.67, 118.33, 118.67, 125]

print("\naxis=1 — one number PER ROW (per stock):")
print(np.mean(portfolio, axis=1))   # [104.25, 51.5, 203.25]

#  check — does axis=1 give you 3 numbers (one per stock)? Does axis=0 give you 4 numbers (one per day)?
````

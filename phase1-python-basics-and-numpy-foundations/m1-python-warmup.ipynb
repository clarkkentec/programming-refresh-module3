**2 - Creating a manual mean function**<br>
*Remembering how functions in python work*

```python

def calculate_mean(numbers):
    # The function calculates the mean (average) of a list of numbers.
    total = 0
    print(f"starting total: {total}")
    print(f"length of numbers (number of elements in your list): {len(numbers)}")
        #len(numbers) is the number of items in the list 
        # Which is used to calculate the mean by dividing the total sum by this "length".
    for num in numbers:
        total += num
        print (f"Current total: {total}")
    return total / len(numbers)



my_numbers = [5, 7, 3, 10, 24]
print(f"Your average is: {calculate_mean(my_numbers)}")
```

**3 - Using Numpy for Arrays**<br>
*Using Numpy functions to do the heavy lifting instead of making a python function*

```python
numpy_mean = np.mean(my_numbers)
print(f"Using numpy, the average is: {numpy_mean}")
#here you can see that the result from our function and the result from numpy's mean function are the same 
# ...which confirms that our function is working correctly.

#Numpy can do so much heavy lifting!
```

**4 - Your first Array**<br>
*Using arrays x Numpy*<br><br>

*Have a think about the differences between lists and arrays*

```python
#Numpy is a powerful library in Python that provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently. It is widely used in scientific computing, data analysis, and machine learning due to its performance and ease of use.

#Establishing a python list and a numpy array to demonstrate the differences between them.
#python_list
python_list = [1, 2, 3, 4, 5]

#numpy_array
numpy_array = np.array(python_list)

#Outputting the python list and the numpy array, along with their types, to show the differences between them.
print(f"Python list: {python_list}")
print(type(python_list))
print(f"Numpy array: {numpy_array}")
print(type(numpy_array))

```

```python
#Playing around with the first element of both the python list and the numpy array to show how they can be accessed in a similar way, but they are different data structures.
print(f"First element of python list: {python_list[0]}")
print(f"First element of numpy array: {numpy_array[0]}")


```

```python
#Now let's see how we can perform operations on both the python list and the numpy array.
print("")
print(python_list*2) # This will repeat the list, not perform element-wise multiplication
print(numpy_array*2) # This will perform element-wise multiplication, multiplying each element by 2


#adding to the python list and the numpy array to show how they can be modified.
print("")
python_list.append(6)
numpy_array = np.append(numpy_array, 6)
print(f"Modified python list: {python_list}")
print(f"Modified numpy array: {numpy_array}")
#the difference in how we modify the python list and the numpy array is evident here. We use the append method for the python list, while for the numpy array we use the np.append function, which creates a new array with the added element.
```

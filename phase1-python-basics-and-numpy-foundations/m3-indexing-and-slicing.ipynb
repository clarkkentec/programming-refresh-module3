In this module, we'll look at Indexing and slicing.<br>This is how you 'reach into' an array and pull out exactly what you want — a single value, a row, a column, or a filtered subset.

```python
import numpy as np
```

**1 - Basic Indexing (of 1D Arrays)**

```python
prices = np.array([101, 103, 99, 107, 110, 108, 115, 113, 120, 118])

#picking out single elements
print(prices[0])  # First element
print(prices[-1])  # Last element
print(prices[-2])  # Second to last element
print("")
#Slicing - picking out a range of elements. 
# The syntax is: array[start:stop:step] - note: the stop index is exclusive, meaning it is not included in the slice.
print("First five elements:\n", prices[0:5])  # First five elements
print("From the sixth element to the end:\n", prices[5:])   # From the sixth element to the end
print("First five elements (same as prices[0:5]):\n", prices[:5])   # First five elements (same as prices[0:5])
print("Every second element:\n", prices[::2])  # Every second element
print("Every second element starting from the second element:\n", prices[1::2])  # Every second element starting from the second element
print("All elements in reverse order:\n", prices[::-1])  # All elements in reverse order
print("Every second element in reverse order:", prices[::-2])  # Every second element in reverse order


```

**2 - 2D Indexing (Matricies)**

```python
#Imagine this as 3 stocks (rows) over 4 days (columns)
portfolio = np.array([[100, 102, 101, 105],     # Stock A
                      [200, 198, 202, 205],     # Stock B
                      [300, 305, 310, 315]])    # Stock C

# Single value - [row, column]
print(portfolio[0, 0])  # Price of Stock A (row 1) on Day 1 (column 1)
print(portfolio[2, 2])  # Price of Stock C (row 3) on Day 3 (column 3)

#Entire row - one stock's prices over all days
print(portfolio[1, :]) # Prices of Stock B (row 2) over all days
print(portfolio[2, :]) # Prices of Stock C (row 3) over all days

#Entire column - all stocks' prices on a specific day
print(portfolio[:, 2]) # Prices of all stocks on Day 3 (column 3)

#slicing - picking out a range of rows and columns
print(portfolio[0:2, 1:3])  # Prices of Stock A and Stock B (rows 1-2) on Day 2 and Day 3 (columns 2-3)
 
# : on its own means "all", so portfolio[:, 1:3] means "all rows, but only columns 1 and 2"
```

**3 - Boolean Indexing**

```python
prices = np.array([101, 103, 99, 107, 110, 108, 115, 113, 120, 118])

mask = prices > 110
print(mask)  # This will print a boolean array where True indicates the condition is met

# Now we can use this boolean array to filter the original prices array
print(prices[prices > 110])  # This will print only the prices that are greater than 110
print(prices[mask])  # This will give the same result as the previous line, using the mask variable

# Combine conditions with & (and) and | (or)
print(prices[(prices > 105) | (prices < 115)])  # between 105 and 115
```

This is powerful because you'll be able to write returns[returns < -0.02] to instantly find all days where your portfolio dropped more than 2%. No loops needed — just one line!

**4 - np.where()‼️ **<br>
*Instead of filtering values, this functions lets you find the indicies (position of the elements/their indexes) where a condition is true*

```python
indicies = np.where(prices > 110)
print(indicies)  # This will print the indices of the elements that are greater than 110
print(indicies[0])  # np.where returns a tuple, so we need to access the first element to get the array of indices

print(prices[indicies])  # This will print the values of the elements that are greater than 110 using the indices
```

Think of boolean indexing as asking "give me the values", and np.where() as asking "give me the positions". Both are useful in different situations!

3. Why does np.diff() return an array that's one element shorter than the input?
<br>
np.diff() returns an array that's one element shorter than the input because the last value does not have a consecutive value to calculate the difference for.<br><br>
Improved answer: It's not the *last* value that has no pair — it's that `np.diff()` computes **between** elements, so for `n` elements there are always `n-1` gaps between them. Like fence posts — 10 posts, 9 gaps! 🪵<br><br>
[101, 103, 99, 107, 110]   ← 5 elements<br>
     ↕    ↕   ↕    ↕       ← 4 differences<br>
[  2,  -4,  8,   3]

## 🚨 The Independence Assumption — Reality Check

### **The Formula Assumes:**
Returns are **i.i.d. (independent and identically distributed)**
- Today's return doesn't influence tomorrow's
- Volatility stays constant
- No correlation between periods

### **Real Markets Violate This In Multiple Ways:**

| Violation | What It Means | Impact on Scaling |
|-----------|---------------|-------------------|
| **Volatility Clustering** | High vol days cluster together ("volatility comes in waves") | Variance scales FASTER than linear → sqrt(252) underestimates annual vol |
| **Momentum/Autocorrelation** | Positive returns tend to follow positive returns (short-term) | Variance scales FASTER than linear |
| **Mean Reversion** | Extreme moves tend to reverse (long-term) | Variance scales SLOWER than linear |
| **Changing Regimes** | Bull markets have different vol than bear markets | Variance isn't stationary — scaling breaks down |

---

## 📊 Real-World Example

**During the 2008 Financial Crisis:**

```python
# Pre-crisis (2006-2007): Daily vol ≈ 0.8%
# Naive annual vol = 0.8% × √252 ≈ 12.7%

# Actual 2008 annual vol ≈ 45% 
# (3.5× higher than scaled estimate!)
```

**Why the massive difference?**
- Volatility clustering: Oct 2008 had daily moves of 10%+
- Returns weren't independent — panic fed on panic
- The sqrt(252) scaling failed catastrophically

## 🎯 So Why Do We Still Use It?

### **Pragmatic Reasons:**

1. **"All models are wrong, but some are useful"**
   - For *normal market conditions*, it's close enough
   - Error from independence violation < error from other sources (parameter estimation, model risk)

2. **Simple and Fast**
   - Alternative: GARCH models (account for vol clustering) — complex, computationally expensive
   - sqrt(252) gives you an instant estimate

3. **Works Well for Daily → Annual (But Not Always)**
   - Daily returns are *less* autocorrelated than hourly/minute data
   - Weekly returns have *more* mean reversion
   - The sweet spot is daily → monthly/annual

4. **Industry Standard**
   - Everyone uses it, so it's the common language
   - When you say "15% annual vol," people know you scaled from daily

---

## ⚠️ When Professionals Know It Breaks

**Good quant analysts:**

### **Don't use sqrt(252) for:**
- **High-frequency data** (minute/hourly) → autocorrelation too strong
- **Crisis periods** → volatility clustering violates independence
- **Long horizons from weekly data** → mean reversion kicks in
- **Options pricing** → use realized/implied vol instead

### **Do use it for:**
- **Quick risk estimates** (daily → annual vol)
- **Backtesting historical strategies** (industry standard comparison)
- **Regulatory reporting** (required by many frameworks)
- **Relative comparisons** (comparing two strategies on same basis)

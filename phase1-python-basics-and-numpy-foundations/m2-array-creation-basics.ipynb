**1 - Core Creation Basics**<br>
*Creating an array and different functions*

```python
import numpy as np
```

```python
# 1. Creating a python list manually
a = np.array([1, 2, 3, 4, 5])

# 2. All 0 - this is useful for initialising empty result arrays
b = np.zeros((3,3))
b2 = np.zeros(5) # this is a 1D array of 5 zeros

# 3. All 1 - same concept, just a different default value
c = np.ones((2,4))

# 4. Arange - this is like python's built in range function, but it returns an array instead of a list)
d = np.arange(0, 10, 2) # this is like range(0, 10, 2) - it starts at 0, goes up to (but not including) 10, and steps by 2

# 5. Linspace - N evenly spaced points between two values
e = np.linspace(0, 1, 10) #this creates an array of 10 evenly spaced points between 0 and 1

# 6. Random numbers between 0 and 1
f = np.random.rand(5) # this creates a 1D array of 5 random numbers between 0 and 1
# 6a. Random integers between a and b
f2 = np.random.randint(0, 10, (3,3)) # this creates a 3x3 array of random integers between 0 and 10 (not including 10

print("np.array:\n", a, "\n")
print("np.zeros:\n", b, "\n")
print("np.zeros:\n", b2, "\n")
print("np.ones:\n", c, "\n")
print("np.arrange:\n", d, "\n")
print("np.linspace:", e, "\n")
print("np.random.rand:\n", f, "\n")
print("np.random.randint:\n", f2, "\n")
```

**2 - Interrogating your arrays**<br>
*Every array has attributes that can be accessed to find out more about the array. This is useful for debugging and understanding your data.*

```python
my_2darray = np.array([[1, 2, 3],[4, 5, 6]])

print("Shape: ", my_2darray.shape,"\n") #(rows, columns) !!
print("Ndim: ", my_2darray.ndim,"\n") # number of dimensions
print("Dtype: ", my_2darray.dtype,"\n") # data type of elements
print("Size: ", my_2darray.size,"\n") # total number of elements (rows * columns for 2D arrays)

#These are called attributes - not functions!! They are properties the array has, not things it does
```

**3 - Arange vs linspace**<br>
*What are the differences between them?*

```python
#Arange - you control the STEP SIZE
# "Give me a range of numbers from 0 to 10, jumping by 2"
print(np.arange(0, 10, 2)) # this is like range(0, 10, 2) - it starts at 0, goes up to (but not including) 10, and steps by 2

#Linspace - you control the NUMBER OF POINTS
# "Give me 10 evenly spaced points between 0 and 10"
print(np.linspace(0, 10, 10)) #this creates an array of 10 !!evenly!! spaced points between 0 and 10
```

💡 When would you use linspace in quant work?<br>*Imagine plotting a smooth price curve or generating time intervals — you know your start, end, and how many points you want. That's linspace territory!*

**4 - Data Type (dtype)**

```python
#Integer Array
int_array = np.array([1,2,3])
print(int_array.dtype)

float_array = np.array([1.0,2.0,3.0])
print(float_array.dtype)

mixed_array = np.array([1,2.0,3])
print(mixed_array.dtype) #numpy promotes to the safer type

```

This automatic promotion is called *type casting* - I.e, numpy picks the type that can safely represent all values.

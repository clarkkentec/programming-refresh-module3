## ❌ Q5: Log Returns — **INCORRECT** ❌

**This is concerning because we covered this extensively in Unit 1!**

```python
# WRONG (your answer):
np.log(prices).sum()  # ❌ This sums all log prices (meaningless!)

# CORRECT:
np.log(prices / prices.shift(1))  # ✅ log(P_t / P_{t-1})

# Or equivalently:
np.log(prices).diff()  # ✅ log(P_t) - log(P_{t-1})
```

**What `np.log(prices).sum()` gives you:**
```
ln(100) + ln(102) + ln(105) + ln(103)
= ln(100 × 102 × 105 × 103)  # Product of all prices!
```

This is NOT log returns!

### 2. Pandas NaN Handling
```python
# By default, Pandas SKIPS NaN in aggregations:
df.mean()      # skipna=True (default)
df.sum()       # skipna=True (default)
df.corr()      # Uses pairwise-complete observations

# To treat NaN as 0:
df.fillna(0).mean()  # Explicit fill first
```

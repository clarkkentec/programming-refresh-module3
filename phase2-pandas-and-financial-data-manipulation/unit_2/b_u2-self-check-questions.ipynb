## Q4: Why Boolean Indexing Preserves Index ❌

**Your answer:**
> "It is because .iloc[] returns a new copy of the array whilst boolean indexing slices the original array"

**This is not quite right.** Both can return views or copies depending on context.

**The real answer:**

**Boolean indexing** selects rows **by their label-based condition**, so it naturally preserves the Index (dates). The result contains the same Index labels as the rows that matched.

**`.iloc[]`** uses **integer positions** (0, 1, 2...) which are independent of Index labels. It treats the DataFrame like a NumPy array — positions, not labels.

**Example:**
```python
# Boolean indexing - preserves meaningful dates
df[df['AAPL'] > 0.02]
# Returns rows with their actual dates: 2024-01-15, 2024-03-22, etc.

# .iloc[] - uses positions
df.iloc[[0, 5, 10]]
# Returns rows 0, 5, 10 but still shows their original Index
# (Index is preserved, but selection was positional, not label-based)
```

**Better answer:** Boolean indexing works with labels (dates, tickers), so the Index is inherently part of the selection logic. `.iloc[]` works with positions (0, 1, 2...) independent of what the Index labels are.

**Grade: 4/10** (wrong reasoning, but you're thinking about views/copies which shows you're engaging with the material)

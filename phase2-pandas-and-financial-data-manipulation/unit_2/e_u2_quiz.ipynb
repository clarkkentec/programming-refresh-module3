## Quick Clarification: .iloc[] Indexing

```python
# Example DataFrame
df = pd.DataFrame(
    {'returns': [0.01, 0.02, 0.03]},
    index=[100, 200, 300]  # Custom Index labels
)

# Using .iloc[] (POSITION-based)
df.iloc[0]  # Returns the FIRST row (label 100)
df.iloc[1]  # Returns the SECOND row (label 200)
df.iloc[2]  # Returns the THIRD row (label 300)

# Using .loc[] (LABEL-based)
df.loc[100]  # Returns the row with label 100
df.loc[200]  # Returns the row with label 200
df.loc[300]  # Returns the row with label 300
```
